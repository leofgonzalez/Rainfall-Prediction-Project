# Importing required libraries

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

# The Data Source

The website chosen for this project was Books to Scrape, you can access it here https://books.toscrape.com/

We'll be looking for the following information about the Books:

- Book Name
- Rating
- Image
- Price

```python
for i in range(1,51):

  url = f"https://books.toscrape.com/catalogue/page-{i}.html"

  response = requests.get(url) # Requesting the information from the website

  response = response.content # Attributing the content to the variable

  soup = BeautifulSoup(response,'html.parser') # Using BeautifulSoup to transform the binary data to HTML

  ol = soup.find('ol') # Finding the Order List / Main group point in the HTML structure to the products (Books)

  articles = ol.find_all('article', class_='product_pod') # Finding each product (Book) in the Order List, each one will have the complete information about each product


# I'll create a list to receive all the information on each product

  books = []

# Each article is organized with the complete information of each product
# Now I'm going to loop into each article and gather the information seeked

  for article in articles:
    image = article.find('img')
    title = image.attrs['alt']
    star = article.find('p')
    price = article.find('p', class_='price_color').text

    star = star['class'][1] # Data Cleaning
    price = float(price[1:]) # Data Cleaning and transformation to numerical content

    books.append([title, price, star])
```

# Transforming the data to CSV - Pandas

```python
df = pd.DataFrame(books, columns=['Title','Price','Star Rating'])

df.to_csv('books.csv')

df
```
